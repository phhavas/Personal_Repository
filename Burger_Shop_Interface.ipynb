```python
# Henrik P. Havas 2026.02.17.
import sys
import os

# implement the classes listed below
class FoodItem:
    name = ""
    amount = int()
    price = float()
    def __init__(self, name, amount, price) -> None:
       self.name = name
       self.amount = amount
       self.price = price
    def display(self):
      print(f"{self.name}")
      print(f"Price: {self.price} Ft")

class Burger(FoodItem):
    sauces = ""
    ingredients = ""
    def __init__(self, name, amount, price, sauces, ingredients) -> None:
       super().__init__(name, amount, price)
       self.sauces = sauces
       self.ingredients = ingredients
    def display(self):
      FoodItem.display(self)
      print(f"Sauces: {self.sauces}")
      print(f"Ingredients: {self.ingredients}\n")

class Drink(FoodItem):
    sugarFree = False
    volume = float()
    def __init__(self, name, amount, price, sugarFree, volume) -> None:
       super().__init__(name, amount, price)
       self.sugarFree = sugarFree
       self.volume = volume
    def display(self):
      FoodItem.display(self)
      print(f"Size: {self.volume} l\n")

class Side(FoodItem):
    size = ""
    def __init__(self, name, amount, price, size) -> None:
       super().__init__(name, amount, price)
       self.size = size
    def display(self):
      FoodItem.display(self)
      print(f"Size: {self.size}\n")

class Combo(FoodItem):
    burger = object()
    side = object()
    drink = object()
    def __init__(self, name, amount, price, burger, side, drink) -> None:
       super().__init__(name, amount, price)
       self.burger = burger
       self.side = side
       self.drink = drink
    def display(self):
      print(self.burger.name)
      print(self.side.name)
      print(self.drink.name)

class Order(FoodItem):
    fullPrice = float()
    burgers = list()
    burger_amount = int()
    sides = list()
    side_amount = int()
    drinks = list()
    drink_amount = int()
    combos = list()
    combo_amount = int()
    def __init__(self, name, amount, price, fullPrice, burgers, burger_amount, sides, side_amount, drinks, drink_amount, combos, combo_amount) -> None:
      super().__init__(name, amount, price)
      self.fullPrice = fullPrice
      self.burger = burgers
      self.burger_amount = burger_amount
      self.side = sides
      self.side_amount = side_amount
      self.drink = drinks
      self.drink_amount = drink_amount
      self.combos = combos
      self.combo_amount = combo_amount
    def add_burger(self, b, b_amount):
      self.burgers.append(b)
      self.burger_amount += b_amount
    def add_side(self, s, s_amount):
      self.sides.append(s)
      self.side_amount += s_amount
    def add_drink(self, d, d_amount):
      self.drinks.append(d)
      self.drink_amount += d_amount
    def add_combo(self, c, c_amount):
      self.combos.append(c)
      self.combo_amount += c_amount
    def add_price(self, p):
      self.fullPrice += p
    def display(self):
      for i in range(len(self.burgers)):
        print(f"{self.burgers[i].name} --- Amount: {self.burger_amount} --- Price: {self.burgers[i].price} Ft")
      for i in range(len(self.sides)):
        print(f"{self.sides[i].name} ----- Amount: {self.side_amount} --- Price: {self.sides[i].price} Ft")
      for i in range(len(self.drinks)):
        print(f"{self.drinks[i].name} ----- Amount: {self.drink_amount} --- Price: {self.drinks[i].price} Ft")
      for i in range(len(self.combos)):
        print(f"{self.combos[i].name}  ----- Amount: {self.combo_amount}")
        print(f"  {self.combos[i].burger.name} --- Price: {self.combos[i].burger.price} Ft")
        print(f"  {self.combos[i].side.name} --- Price: {self.combos[i].side.price} Ft")
        print(f"  {self.combos[i].drink.name} --- Price: {self.combos[i].drink.price} Ft")
      print(f"Full price: {self.fullPrice} Ft")

# MENU
# Burgers
classicCheese = Burger("Classic Cheeseburger",1,780,"ketchup, mustard","burger bun, beef patty, cheddar cheese, pickles, onion")
baconBBQ = Burger("Bacon BBQ Burger",1,750,"BBQ sauce","burger bun, beef patty, cheddar cheese, onion, pickles")
spicyBeef = Burger("Spicy Beefburger",1,810,"spicy sauce", "burger bun, beef patty, iceberg lettuce, fried onions")

# Sides
frenchFriesSmall = Side("French Fries (Small)",1,780,"S")
frenchFriesMedium = Side("French Fries (Medium)",1,1180,"M")
frenchFriesLarge = Side("French Fries (Large)",1,1270,"L")
onionRings = Side("Onion Rings (5 pcs)",1,690,"M")
nuggetsSmall = Side("Chicken Nuggets (4 pcs)",1,960,"S")
nuggetsMedium = Side("Chicken Nuggets (6 pcs)",1,1430,"M")
nuggetsLarge = Side("Chicken Nuggets (9 pcs)",1,1730,"L")

# Drink
cocaCola = Drink("Coca-Cola",1,750,False,0.25)
spriteZero = Drink("Sprite Zero",1,740,True,0.25)
fantaZero = Drink("Fanta Orange Zero",1,740,True,0.25)

# Display menu function
def display_menu():
  print("\nBURGERS:")
  classicCheese.display()
  baconBBQ.display()
  spicyBeef.display()
  print("SIDES:")
  frenchFriesSmall.display()
  frenchFriesMedium.display()
  frenchFriesLarge.display()
  onionRings.display()
  nuggetsSmall.display()
  nuggetsMedium.display()
  nuggetsLarge.display()
  print("DRINKS:")
  cocaCola.display()
  spriteZero.display()
  fantaZero.display()

def user_input_burger():
    # ask user for input and store it in burger object
    while True:
      burger = input("Please select a burger from the menu: ").capitalize()

      # Check if the entered input is on the burger menu
      if burger in ["Classic cheeseburger","Bacon bbq burger","Spicy beefburger"]:
        break
      elif burger == "Cancel":
        print("Thank you!")
        sys.exit(0)
      else:
        print("Sorry, this item is not on the menu. The following burgers are available: ")
        classicCheese.display()
        baconBBQ.display()
        spicyBeef.display()

    # Ask the user for amount
    amount = input("How many burgers would you like? ").lower()
    if amount == "cancel":
      print("Thank you!")
      sys.exit(0)

    # Convert to integer
    amount = int(amount)

    # Create object according to user input
    if burger == "Classic cheeseburger":
      b = Burger("Classic Cheeseburger",amount,amount*780,"ketchup, mustard","burger bun, beef patty, cheddar cheese, pickles, onion")
    elif burger == "Bacon bbq burger":
      b = Burger("Bacon BBQ Burger",amount,amount*750,"BBQ sauce","burger bun, beef patty, cheddar cheese, onion, pickles")
    elif burger == "Spicy beefburger":
      b = Burger("Spicy Beefburger",amount,amount*810,"spicy sauce","burger bun, beef patty, iceberg lettuce, fried onions")
    return b

def user_input_drink():
    # ask user for input and store it in drink object
    while True:
      drink = input("Please select a drink from the menu: ").capitalize()

      # Check if the entered input is on the burger menu
      if drink in ["Coca-cola","Sprite zero","Fanta orange zero"]:
        break
      elif drink == "Cancel":
        print("Thank you!")
        sys.exit(0)
      else:
        print("Sorry, this item is not on the menu. The following drinks are available: ")
        cocaCola.display()
        spriteZero.display()
        fantaZero.display()

    # Ask the user for amount
    amount = input("How many drinks would you like? ").lower()
    if amount == "cancel":
      print("Thank you!")
      sys.exit(0)

    # Convert to integer
    amount = int(amount)

    # Create object according to user input
    if drink == "Coca-cola":
      d = Drink("Coca-Cola",amount,amount*750,False,0.25)
    elif drink == "Sprite zero":
      d = Drink("Sprite Zero",amount,amount*740,True,0.25)
    elif drink == "Fanta orange zero":
      d = Drink("Fanta Orange Zero",amount,amount*740,True,0.25)
    return d

def user_input_side():
    # ask user for input and store it in side object
    while True:
      side = input("Please select a side from the menu: ").capitalize()

      # Check if the entered input is on the burger menu
      if side in ["French fries (small)","French fries (medium)","French fries (large)","Onion rings (5 pcs)","Chicken nuggets (4 pcs)","Chicken nuggets (6 pcs)","Chicken nuggets (9 pcs)"]:
        break
      elif side == "Cancel":
        print("Thank you!")
        sys.exit(0)
      else:
        print("Sorry, this item is not on the menu. The following sides are available: ")
        frenchFriesSmall.display()
        frenchFriesMedium.display()
        frenchFriesLarge.display()
        onionRings.display()
        nuggetsSmall.display()
        nuggetsMedium.display()
        nuggetsLarge.display()

    # Ask the user for amount
    amount = input("How many sides would you like? ").lower()
    if amount == "cancel":
      print("Thank you!")
      sys.exit(0)

    # Convert to integer
    amount = int(amount)

    # Create object according to user input
    if side == "French fries (small)":
      s = Side("French Fries (Small)",amount,amount*780,"S")
    elif side == "French fries (medium)":
      s = Side("French Fries (Medium)",amount,amount*1180,"M")
    elif side == "French fries (large)":
      s = Side("French Fries (Large)",amount,amount*1270,"L")
    elif side == "Onion rings (5 pcs)":
      s = Side("Onion Rings (5 pcs)",amount,amount*690,"M")
    elif side == "Chicken nuggets (4 pcs)":
      s = Side("Chicken Nuggets (4 pcs)",amount,amount*960,"S")
    elif side == "Chicken nuggets (6 pcs)":
      s = Side("Chicken Nuggets (6 pcs)",amount,amount*1430,"M")
    elif side == "Chicken nuggets (9 pcs)":
      s = Side("Chicken Nuggets (9 pcs)",amount,amount*1730,"L")
    return s

def user_input_combo():
    # ask user for input and store it in combo object
    # a combo must include one burger, one side, and one drink

    # Asking user to select the burger
    while True:
      burger = input("Please select a burger from the menu: ").capitalize()

      # Check if entered burger is on the menu
      if burger in ["Classic cheeseburger","Bacon bbq burger","Spicy beefburger"]:
        break
      elif burger == "Cancel":
        print("Thank you!")
        sys.exit(0)
      else:
        print("Sorry, this item is not on the menu. The following burgers are available: ")
        classicCheese.display()
        baconBBQ.display()
        spicyBeef.display()

    # Create burger object
    if burger == "Classic cheeseburger":
      b = Burger("Classic Cheeseburger", 1, 780,"ketchup, mustard","burger bun, beef patty, cheddar cheese, pickles, onion")
    elif burger == "Bacon bbq burger":
      b = Burger("Bacon BBQ Burger", 1, 750,"BBQ sauce","burger bun, beef patty, cheddar cheese, onion, pickles")
    elif burger == "Spicy beefburger":
      b = Burger("Spicy Beefburger", 1, 810,"spicy sauce","burger bun, beef patty, iceberg lettuce, fried onions")

    # Asking user to select the side
    while True:
      side = input("Please select a side from the menu: ").capitalize()

      # Check if entered side is on the menu
      if side in ["French fries (small)","French fries (medium)","French fries (large)","Onion rings (5 pcs)","Chicken nuggets (4 pcs)","Chicken nuggets (6 pcs)","Chicken nuggets (9 pcs)"]:
        break
      elif side == "Cancel":
        print("Thank you!")
        sys.exit(0)
      else:
        print("Sorry, this item is not on the menu. The following sides are available: ")
        frenchFriesSmall.display()
        frenchFriesMedium.display()
        frenchFriesLarge.display()
        onionRings.display()
        nuggetsSmall.display()
        nuggetsMedium.display()
        nuggetsLarge.display()

    # Create side object
    if side == "French fries (small)":
      s = Side("French Fries (Small)",1,780,"S")
    elif side == "French fries (medium)":
      s = Side("French Fries (Medium)",1,1180,"M")
    elif side == "French fries (large)":
      s = Side("French Fries (Large)",1,1270,"L")
    elif side == "Onion rings (5 pcs)":
      s = Side("Onion Rings (5 pcs)",1,690,"M")
    elif side == "Chicken nuggets (4 pcs)":
      s = Side("Chicken Nuggets (4 pcs)",1,960,"S")
    elif side == "Chicken nuggets (6 pcs)":
      s = Side("Chicken Nuggets (6 pcs)",1,1430,"M")
    elif side == "Chicken nuggets (9 pcs)":
      s = Side("Chicken Nuggets (9 pcs)",1,1730,"L")

    # Asking user to select the drink
    while True:
      drink = input("Please select a drink from the menu: ").capitalize()

      # Check if entered drink is on the menu
      if drink in ["Coca-cola","Sprite zero","Fanta orange zero"]:
        break
      elif drink == "Cancel":
        print("Thank you!")
        sys.exit(0)
      else:
        print("Sorry, this item is not on the menu. The following drinks are available: ")
        cocaCola.display()
        spriteZero.display()
        fantaZero.display()

    # Create drink object
    if drink == "Coca-cola":
      d = Drink("Coca-Cola",1,750,False,0.25)
    elif drink == "Sprite zero":
      d = Drink("Sprite Zero",1,740,True,0.25)
    elif drink == "Fanta orange zero":
      d = Drink("Fanta Orange Zero",1,740,True,0.25)

    # Calculating full price of the combo
    fullPrice = b.price + s.price + d.price

    # Create combo object
    c = Combo("Combo",1,fullPrice, b, s, d)
    return c

def take_order():
    # ask user for name for the order
    # repeat taking order until client is done
    # display order details
    # display a thank you message

    # Print welcome message
    print("Welcome to Jack's Burger Shop\n")

    # Ask the user for their name
    name = input("Please enter your name: ").capitalize()
    if name == "Cancel":
      print("Thank you!")
      return

    # Display menu
    display_menu()

    # Initialize variable for while loop
    anything_else = "yes"

    # Initialize order variable
    o = Order("Total Order", 0, 0, 0, burgers=[], burger_amount=0, sides=[], side_amount=0, drinks=[], drink_amount=0, combos=[], combo_amount=0)

    # Ask the user for the order inputs
    while(anything_else == "yes"):
      while True:

        # Check if the user input is appropriate
        items = input("Do you want a burger, a side, a drink, or a combo? ").lower()
        if items in ["burger","side","drink","combo"]:
          break
        elif items == "cancel":
          print("Thank you!")
          return
        else:
          print(f"Sorry, I can't recognise {items}. Please type in burger, side, drink, or combo")

      # Create items according to user input
      if items == "burger":
        burger = user_input_burger()
        o.add_price(burger.price)
        o.add_burger(burger, burger.amount)
      elif items == "side":
        side = user_input_side()
        o.add_price(side.price)
        o.add_side(side, side.amount)
      elif items == "drink":
        drink = user_input_drink()
        o.add_price(drink.price)
        o.add_drink(drink, drink.amount)
      elif items == "combo":
        combo = user_input_combo()
        o.add_price(combo.price)
        o.add_combo(combo, combo.amount)

      # Ask the user if they want anything else
      anything_else = input("Would you like to order anything else? ").lower()
      if anything_else == "cancel":
        print("Thank you!")
        return

    # Display the full order
    o.display()

    # Thank you message
    print("Thank you! Your order is being prepared.")

take_order()
```
